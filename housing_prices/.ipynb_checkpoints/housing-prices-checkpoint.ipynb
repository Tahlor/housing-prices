```python
def prep_features(features, targets):
    # Training test split, normalize
    from sklearn.preprocessing import StandardScaler
    from sklearn.model_selection import train_test_split

    # Remove features
    features.drop(["Id"], axis=1)

    # Fill NA
    #features.fillna()

    X_train, X_test, y_train, y_test = train_test_split(features, targets, train_size=0.8, random_state=42)

    display(X_train)

    # Normalize
    scaler = StandardScaler().fit(X_train)
    X_train_scaled = pd.DataFrame(scaler.transform(X_train), index=X_train.select_dtypes(include=[np.number]), columns=X_train.columns.values)
    X_test_scaled = pd.DataFrame(scaler.transform(X_test), index=X_test.select_dtypes(include=[np.number]), columns=X_test.columns.values)
    return X_train_scaled, y_train, X_test_scaled, y_test
```

```python
def execute_model(X_train, y_train, X_test, y_test):
    from sklearn.ensemble import RandomForestRegressor
    rf = RandomForestRegressor(n_estimators=500, oob_score=True, random_state=0)
    rf.fit(X_train, y_train)

    from sklearn.metrics import r2_score
    from scipy.stats import spearmanr, pearsonr
    predicted_train = rf.predict(X_train)
    predicted_test = rf.predict(X_test)
    test_score = r2_score(y_test, predicted_test)
    spearman = spearmanr(y_test, predicted_test)
    pearson = pearsonr(y_test, predicted_test)

    print('Out-of-bag R-2 score estimate: {}'.format(rf.oob_score_))
    print('Test data R-2 score: {}'.format(test_score))
    print('Test data Spearman correlation: {}'.format(spearman[0]))
    print('Test data Pearson correlation: {}'.format(pearson[0]))
```

```python
def main(path_training, path_test, training_test_seed=None, output_path = r"./output"):
    """Processes data, predicts housing prices, saves predictions.

    Args:
        path_training  (str): Path to training data.
        path_test (str): Path to test data.
        training_test_seed (int, optional): Random seed for internal training/test split.
        output_path ( str): Location to save predicitions

    Returns:
        None
    """


    training_df = import_data.process_data(path_training)
    features = pd.get_dummies(training_df)
    with pd.option_context('display.max_rows', 5, 'display.max_columns', None):
        display(features)
    y = features['SalePrice']
    X_train, y_train, X_test, y_test = prep_features(features, y)
    execute_model(X_train, y_train, X_test, y_test)
```
